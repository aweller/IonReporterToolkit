```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import scipy.stats

project = "quasar40_noQC_vs_QC2"
```

```python
mutsig_genes = ['APC',
 'PIK3CA',
 'EGFR',
 'C11orf30',
 'PTEN',
 'FBXW7',
 'SMAD4',
 'BRAF',
 'TP53',
 'HRAS',
 'JAK3',
 'STK11',
 'KDR',
 'NF1',
 'ERBB2',
 'RUNX1',
 'ATM',
 'TET2',
 'ATRX',
 'MTOR',
 'PTPN11',
 'MAP2K1',
 'ERBB4',
 'AR',
 'HDAC1',
 'CYP2D6',
 'MLH1',
 'SMARCB1']

effect_order = ["MODIFIER","LOW","MODERATE","HIGH"]
```

```python
filename = "/home/andreas/bioinfo/projects/wtc_quasar_analysis/data/quasar_vcfs_default_parameters_noQC/all_samples_default_parameters_noQC_noCNVs_annotated.tsv" # all variants, no QC except removing CNV entries
raw = pd.read_csv(filename, sep = "\t")
raw = raw[raw.alt != "<CNV>"]
```

```python
df_dict = {"sample" : raw["sample"],
           "chrom" : raw["chrom"],
           "pos" : raw["pos"],
           "ref" : raw["ref"],
           "alt" : raw["alt"],
           "fao" : raw["FAO"],
           "fdp" : raw["FDP"],
           "qual" : raw["GQ"],
           "rsid" : raw["snp135"],
            "vartype" : raw["TYPE"],
            "effect" : raw["Effect"],
            "impact" : raw["Effect_Impact"],
            "gene" : raw["Gene_Name"]}

df = pd.DataFrame(df_dict)
```

```python
print "Samples:", df.sample.nunique()
print "Total variants:", len(df)
```

```python
df["chrompos"] = df.chrom + "\t" + df.pos.astype(str)
df = df.drop(["chrom", "pos"], axis = 1)
df["vartype"] = df.apply( lambda x: x["vartype"].split(",")[0], axis = 1)
df["mut"] = df.ref + df.alt
df["dp"] = df.fao.astype(int) + df.fdp.astype(int)
```

```python
# parse and remove the low confidence positions per samples based on normal coverage

filename = "/home/andreas/bioinfo/projects/wtc_quasar_analysis/data/quasar_vcfs_default_parameters_noQC/low_confidence_positions.txt"
lowconf_variants = [row.strip() for row in open(filename).readlines()]

df["lowconf"] = df.apply(lambda x: x["sample"] +"\t"+ x["chrompos"] in lowconf_variants, axis= 1)
df = df[~df.lowconf]
```

```python
df["fao"] = df["fao"].astype(float)
df["fdp"] = df["fdp"].astype(float)
df["recurrence"] = df.apply(lambda x: len( df[df.chrompos == x["chrompos"]] ), axis =1 )
```

```python
def get_allelfreq(row):
    fao = row["fao"]
    fdp = row["fdp"]
    
    return fao/fdp

df["allelfreq"] = df.apply(get_allelfreq, axis = 1)

df["allelfreq_cutoff"] = df.apply(lambda x: x["allelfreq"] > 0.05, axis = 1)
df["dp_cutoff"] = df.apply(lambda x: x["dp"] > 400, axis = 1)
df["mutsig_gene"] = df.gene.isin(mutsig_genes)
df["has_rsid"] = df.rsid != "-"

myfunc = lambda x: True if ((x["dp_cutoff"]) & (x["allelfreq_cutoff"])) else False
df["pass_QC2"] = df.apply(myfunc, axis =1)
```

```python
print "Samples:", df.sample.nunique()

print "Total variants:", len(df)
print "Variants after QC2 cutoff:", len(df[(df.dp > 400) & (df.allelfreq > 0.05)])
```

## Parse sequence information

```python
header_names = ["chr", "pos", "seq", "gcc"]
seq_filename = "/home/andreas/datadump/quasar_variants_seqs.txt"

seqs = pd.read_csv(seq_filename, sep = "\t", header = None, names = header_names)
seqs["chrompos"] = seqs.chr.astype(str) + seqs.pos.astype(str)
```

```python
seqs.head()
```

```python
def find_homopolymers(row):
    seq = row["seq"]
    center = seq[5:16]
        
    sizes = [0,]
    
    for base in ["A", "C", "T", "G"]:
        for size in range(10,0,-1):
            run = base*size
            if run in center:
                sizes.append(size)
    sizes.sort()
    homo = sizes[-1]
    if homo > 5: 
        homo = 5

    return homo

#seqsh = seqs.head()
#seqsh["homo"] = seqsh.apply(find_homopolymers, axis=1)
#print seqsh

seqs["homo"] = seqs.apply(find_homopolymers, axis=1)
homo = {x:y for x,y in zip(seqs.chrompos, seqs.homo)}
df["homo"] = df.chrompos.apply(lambda x: homo.get(x,NaN))
```

```python
print "Total vars:", len(df)

print "Vars failing QC2:", len(df[~df.pass_QC2])
print "Vars passing QC2:", len(df[df.pass_QC2])
print "Unique vars passing QC2:", len( df[df.pass_QC2].drop_duplicates(cols = "chrompos") )

print "Vars passing QC2 outside sign:", len(df[df.pass_QC2 & (~df.mutsig_gene)])
print "Vars passing QC2 inside sign:", len(df[df.pass_QC2 & df.mutsig_gene])
print "Vars passing QC2 inside sign with strong effect:", len(df[df.pass_QC2 & df.mutsig_gene & (df.impact.isin(["HIGH"]))])
print "Positions passing QC2 inside sign with strong effect:", len(df[df.pass_QC2 & df.mutsig_gene & (df.impact.isin(["HIGH"]))].drop_duplicates(cols=["chrompos"]))

print "Median sample count failing QC2:", df[~df.pass_QC2].recurrence.median()
print "Median sample count passing QC2:", df[df.pass_QC2].recurrence.median()
```
